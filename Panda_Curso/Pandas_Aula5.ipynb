```python
import pandas as pd

valores_ausentes = ["**", "###!", "####", "*****", "****", "NULL"]

arquivo = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values = valores_ausentes)

arquivo_dataframe = pd.DataFrame(arquivo)

arquivo_dataframe.head(50)

# filtro_uf = arquivo.ocorrencia_uf.isnull()

# filtro_uf.value_counts()

```

```python
filtro_aerodromo = arquivo.ocorrencia_aerodromo.isnull()

filtro_aerodromo.value_counts()
```

```python
#Filtro que ir√° criar um dataframe com as cidades onde as recomenda√ß√µes s√£o maiores que 5
filtro_recomendacoes = arquivo.total_recomendacoes >= 5
arquivo.loc[filtro_recomendacoes, ["ocorrencia_cidade", "total_recomendacoes"]]
```

```python
#Agora irei criar um filtro para buscar os incidentes graves no estado do Rio de janeiro
filtro_ocorrencia = arquivo.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro_estado = arquivo.ocorrencia_uf == 'RJ'
arquivo.loc[filtro_ocorrencia & filtro_estado] #iremos criar um dataframe com as informa√ß√µes dos incidentes graves no Rio de Janeiro 
```

```python
#Agora irei criar um filtro para buscar os incidentes graves no estado do Rio de janeiro
filtro_ocorrencia = arquivo.ocorrencia_classificacao == 'INCIDENTE'
filtro_estado = arquivo.ocorrencia_uf == 'RJ'
arquivo.loc[filtro_ocorrencia | filtro_estado]
```

```python
# O metodo isin retorna um dataframe com True ou False caso dos dados que queremos estejam contidos no dataframe
# Em nosso caso sempre ser√° verdadeiro(True) quando os dados forem INCIDENTES GRAVES e INCIDENTE 
filtro_isin = arquivo.ocorrencia_classificacao.isin(["INCIDENTE GRAVE", "INCIDENTE"])
print(filtro_isin)
```
